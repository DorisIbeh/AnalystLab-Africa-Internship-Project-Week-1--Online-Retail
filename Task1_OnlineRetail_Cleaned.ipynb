```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\USER\OneDrive\Desktop\OnlineRetail.csv", encoding= 'latin1')
df
```

```python
#converting InvoiceDate column into Datetime format, replace invalid dates with NaT

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
df
```

```python
#Extract Year, MonthName, Day from the InvoiceDate column for time base analysis

df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
df['MonthName'] = df['InvoiceDate'].dt.month_name() # Jan, Dec 
df
```

```python
#Show data type for each column

df.info()
```

```python
# Display numerical features and categorical features

numerical = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical = df.select_dtypes(include=['object']).columns.tolist()
print("Numerical:", numerical) #["Quantity', 'UnitPrice', 'CustomerID', 'Year', 'Month', 'Day']
print("Categorical:", categorical) #['InvoiceNo', 'StockCode', 'Description', 'Country']
```

```python
# Display the number of unique values in each column
df.nunique()
```

# Brief Description of the Dataset
Answer: The Online Retail dataset contains UK e-commerce transaction data from December 2010 to December 2011, its contains 12 Columns which includes
(InvoiceNo,StockCode,Description, Quantity, InvoiceDate, UnitPrice,CustomerID,Country,Year,MonthName,Day,Month) and 541909 Rows












```python
# Replace missing values with blank string
df.fillna('')
```

```python
# Remove duplicate records from the dataset
df.drop_duplicates()
```

```python
df
```

```python
# Remove invalid Quantity and UnitPrice. Negatives = cancellations/returns
print("Negative Qty:", (df['Quantity'] <= 0).sum()) 
print("Negative Price:", (df['UnitPrice'] <= 0).sum())

df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

print(df.shape)

```

```python
# Remove negative signs and convert UnitPrice to float data type
df.loc[:,'UnitPrice'] = df['UnitPrice'].astype(str).str.lstrip('-').astype(float)
# Remove negative signs and convert Quantity to float data type
df.loc[:,'Quantity'] = df['Quantity'].astype(str).str.lstrip('-').astype(float)
df
```

```python
# Remove unwanted leading and trailing characters from product descriptions
df.loc[:,'Description'] = (
    df['Description']
        .str.strip()
        .str.lower()
        .str.title()
)
df
```

```python
# Generate Summary Statistics for numerical columns and Exporatory Data Pattern
```

```python
# Generate summary statistics for sales quantity and unit price
num_cols = ['Quantity', 'UnitPrice']
summary =df[num_cols].agg(['mean','median', 'max', 'std']).round(2)
summary
```

```python
# Create a Revenue column by multiplying quantity sold by unit price
df.loc[:,'Revenue'] = df['Quantity'] * df['UnitPrice']
df['Revenue'].head()   ## A new revenue metric was engineered to measure the monetary value of each transaction,
# enabling sales performance analysis across products, customers, and countries.
```

```python
df
```

```python
# Metrics
# 1. Total Revenue
total_revenue = df['Revenue'].sum()
print(f"Total Revenue: Â£{total_revenue:,.2f}")

# 2. Unique Customers
unique_customers = df['CustomerID'].nunique()
print(f"Unique Customers: {unique_customers}")

# 3. Unique Products
unique_products = df['StockCode'].nunique()
# If you want to count by product name instead:
# unique_products = df['Description'].nunique()
print(f"Unique Products: {unique_products}")

# 4. Countries Served
countries_served = df['Country'].nunique()
print(f"Countries Served: {countries_served}")

# 5. Avg Order Value
aov = total_revenue / df['InvoiceNo'].nunique()
print(f"Avg Order Value: Â£{aov:,.2f}")
```

```python
## exploratory Analysis for E-Commerce Dataset
```

```python
# identify and virtualize the top 10 -selling products by quantity

top_products_qty = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_products_qty.values, y=top_products_qty.index, orient='h')
plt.title('Top 10 Products by Quantity', fontsize=14,weight='bold')
plt.xlabel('Total Units Sold')
plt.ylabel('Product Description')    
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```

```python
top_products_qty
```

```python
# Insight- The best selling products by quantity are low-price, and are not high revenue drivers. This shows that there is heavy reliance;
# on fast moving , low price items. In terms of quantity sold, items like WORLD WAR 2 GLIDERS ASSTD DESIGN lead significantly.
```

```python
# Identify and visualize the top 10 highest revenue-generating countries

df['Revenue'] = df['Quantity'] * df['UnitPrice']
top_countries_rev = df.groupby('Country')['Revenue'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(y=top_countries_rev.values, x=top_countries_rev.index, palette='Reds_r', hue =top_countries_rev.index,legend = False)
plt.title('Top 10 Countries by Total Revenue', fontsize=14, weight='bold')
plt.xlabel('Total Revenue (Â£)')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
top_countries_rev
```

```python
#Insight - This plot shows that revenue is highly concentrated in the UK with a steep drop off to other countries markets.
```

```python
# Identify Monthly sales trend (DEC 2010-DEC 2011)

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Revenue'] = df['Quantity'] * df['UnitPrice'] 

monthly_sales = df.set_index('InvoiceDate').resample('ME')['Revenue'].sum()
plt.figure(figsize=(12,5))
monthly_sales.plot(marker='o', linewidth=2)
plt.title('Monthly Sales Trend (Â£)', fontsize=14, weight='bold')
plt.xlabel('Month')
plt.ylabel('Total Revenue (Â£)')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
monthly_sales
```

```python
#Key Insight- Sales are highly seasonal , Q4 (sept, oct, nov drives the most annual revenue).
# there is a big peak in nov-dec because thats christmas and gifting season, retailers stock up on decor items.
#Sales trend analysis helps identify peak revenue periods, seasonal demand patterns, and fluctuations in customer purchasing behavior.
```

```python
# identify the Most Purchased Products

top_products_count = df['Description'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_products_count.values, y=top_products_count.index, color = 'green')
plt.title('Top 10 Most Purchased Products by Invoice Count', fontsize=14, weight='bold')
plt.xlabel('Number of Invoices')
plt.ylabel('Product')
plt.tight_layout()
plt.show()
```

```python
top_products_count
```

```python
#Insight- The chart shows .that small number of products dominate purchase frequency,
# with items Like WHITE HANGING HEART T-LIGHT HOLDER AND REGENCY CAKESTAND 3 TIER leading significantly.
```

```python
# Identify the top 5 customers by total spend

df['Revenue'] = df['Quantity'] * df['UnitPrice']

customer_spend = df.groupby('CustomerID')['Revenue'].sum()

top_5_customers = customer_spend.sort_values(ascending=False).head(5)

avg_spend_per_customer = customer_spend.mean() 
median_spend_per_customer = customer_spend.median() # better if data is skewed

print(f"Average Spend per Customer: Â£{avg_spend_per_customer:.2f}")
print(f"Median Spend per Customer:  Â£{median_spend_per_customer:.2f}")
print("\nTop 5 Customers by Total Spend:")
print(top_5_customers)

plt.figure(figsize=(10,6))
sns.barplot(y=top_5_customers.values, x=top_5_customers.index.astype(str), hue = top_5_customers.index.astype(str), legend= False)
plt.title('Top 5 Customers by Total Spend (Â£)', fontsize=14, weight='bold')
plt.xlabel('Total Revenue (Â£)')
plt.ylabel('CustomerID')
plt.tight_layout()
plt.show()
```

```python
#Insight-  A small number of customers contribute a significant share of total revenue, highlighting the importance of customer retention and relationship management.
```
